```python
import pandas as pd
import numpy as np
```

```python
import os

current_dir = os.path.dirname(os.path.abspath("__file__"))
file_path = os.path.join(current_dir, "fund_basic_info.csv")
df = pd.read_csv(file_path, dtype=str, keep_default_na=False)
df
```

```python
file_path = os.path.join(current_dir, "fund_manager_em.csv")
df_2 = pd.read_csv(file_path, dtype=str, keep_default_na=False)
df_2
```

```python
df_merged = pd.merge(df, df_2, left_on='基金简称', right_on='现任基金', how='inner')
df_merged
```

```python
df_merged = df_merged[["基金代码", "基金简称", "基金类型", "姓名", "所属公司","现任基金", "累计从业时间","现任基金资产总规模","现任基金最佳回报"]]

df_merged.columns = ['fund_code', 'fund_name', 'fund_type','manager_name', 'company', 'current_fund', 'experience_days', 'total_asset', 'best_return']

df_merged.to_csv('fund_manager.csv', index=False)
```

```python
def join_unique_preserve_order(series, sep):
    seen = set()
    result = []
    for item in series:
        if item not in seen:
            seen.add(item)
            result.append(item)
    return sep.join(result)

df_merged_unique = df_merged.groupby(['fund_code', 'fund_name'], as_index=False).agg({
    'fund_type': lambda x: join_unique_preserve_order(x, '/ '),
    'manager_name': lambda x: join_unique_preserve_order(x, '/ '),
    'company': lambda x: join_unique_preserve_order(x, ','),
    'current_fund': lambda x: join_unique_preserve_order(x, '/ '),
    'experience_days': lambda x: join_unique_preserve_order(x, '/ '),
    'total_asset': lambda x: join_unique_preserve_order(x, '/ '),
    'best_return': lambda x: join_unique_preserve_order(x, '/ ')
})
df_merged_unique
```

```python
df_merged_unique.to_csv('fund_manager.csv', index=False)
```
